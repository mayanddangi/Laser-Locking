```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
percent=50.0
TimePeriod=1.0
Cycles=20
dt=0.01 

t=np.arange(0,Cycles*TimePeriod,dt); 
pwm= (t%TimePeriod<TimePeriod*percent/100)*5
plt.plot(t,pwm)
```

```python
fc = 0.001
alpha = 2*np.pi*dt*fc/(2*np.pi*dt*fc+1)
```

```python
alpha = 0.000001
y = [2.5]
for i in pwm:
    y_t = alpha*i + (1.0-alpha)*y[-1]
    y.append(y_t)
plt.plot(t, y[1:])
```
